## 3. Function to calculate Taxi Fare

Define a function called `get_fare(distance)` that calculates a taxi fare based on the following:
- Base fare is $5.00.
- If distance is greater than 10km, add $1.50 per km for every km **above** 10.
- Return the total fare.

Make use of your function to print 

The fare for 15km is $12.50.

```python
# Type your code here:
def get_fare(distance):
    base_fare = 5.00
    if distance > 10:
        total = base_fare + (distance-10) * 1.5
    else:
        total = base_fare
    return total

print(f"The fare for 15km = {get_fare(15):.2f}")
```
